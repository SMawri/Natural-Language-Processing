## BagOfWords Implementation

```python
# Cleaning the Texts
import re
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
from nltk.stem import WordNetLemmatizer
```

```python
# Stemming
ps=PorterStemmer()
wordnet=WordNetLemmatizer()
corpus=[]
for i in range(len(sentences)):
    review=re.sub('[^a-zA-Z]',' ',sentences[i]) ## Cleaning the sentences removing punctuation.
    review=review.lower()
    review=review.split() ## after split will have a list of words
    review=[ps.stem(word) for word in review if not word in set(stopwords.words('english'))]
    review=" ".join(review)
    corpus.append(review)
```

```python
# Lemmatization 
ps=PorterStemmer()
wordnet=WordNetLemmatizer()
lemmatize_corpus=[]
for i in range(len(sentences)):
    review=re.sub('[^a-zA-Z]',' ',sentences[i]) ## Cleaning the sentences removing punctuation.
    review=review.lower()
    review=review.split() ## after split will have a list of words
    review=[wordnet.lemmatize(word) for word in review if not word in set(stopwords.words('english'))]
    review=" ".join(review)
    lemmatize_corpus.append(review)
```

```python
# Creating a Bag Of Words Model
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=1500)
X=cv.fit_transform(corpus).toarray()
```

```python
# Creating a Bag Of Words Model
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=1500)
X_L=cv.fit_transform(lemmatize_corpus).toarray()
```
